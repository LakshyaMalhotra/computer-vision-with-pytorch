```python
# imports
import os

import pandas as pd
```

```python
# define paths
base_path = os.path.dirname(os.getcwd())
data_dir = "data/describable_textures_dtd"
data_path = os.path.join(base_path, data_dir)
```

```python
# load class names and ids CSV file
df = pd.read_csv(os.path.join(data_path, "class_names_ids.csv"))
df.head()
```

```python
# find all the images in the class name dirs
image_names = {}
for _, row in df.iterrows():
    name = row.class_name
    images = os.listdir(os.path.join(data_path, f"images/{name}"))
    images = [img for img in images if img.endswith(".jpg")]
    print(f"Class {name} has {len(images)} images.")
    image_names[name] = images
```

```python
# create new dataframe with image file names as a separate column
image_df = pd.DataFrame(data=image_names)
image_df = pd.DataFrame(image_df.unstack()).reset_index(level=[0, 1]).drop(columns=["level_1"])
image_df.columns = ["class_name", "file_name"]
image_df
```

```python
# merge the original data frame with the new one
df = df.merge(image_df, left_on="class_name", right_on="class_name")
df.head()
```

```python
# save dataframe
df.to_csv(os.path.join(data_path, "data.csv"), index=False)
```
